Reverse a given string without using `[::-1]` or `reversed()`.

```python
myStr = "ayan"
res = ""
for i in myStr:
    res = i + res
print(res)
```

Count the number of vowels in a string.

```python
def vowelCounter(n):
    vList = ["a","e","i","o","u"]
    c = 0
    for i in n:
        if i in vList:
            c += 1
    return c

print(vowelCounter("abcdefghijklmnopqrstuvwxyz"))
```

Check if a string is a palindrome

```python
myStr = "ava"
if myStr == myStr[::-1]:
    print("Palindrome")
else:
    print("No")
```

Remove all spaces from a string

```python
myStr = input("Enter your string: ")
strWOSpace = myStr.replace(" ", "")
print(strWOSpace)
```

Capitalize the first letter of each word in a sentence.

```python
myStr = "i stay in bangalore"
modifiedStr = myStr.title()
print(modifiedStr)
```

Find the longest word in a sentence.

```python
myStr = "cow drinks water"
splittedLsit = myStr.split() 

longestWord = max(splittedLsit, key=len)

print(longestWord)
```

Check if two strings are anagrams.

```python
myStr1 = "ayan"
myStr2 = "naya"
if sorted(myStr1) == sorted(myStr2):
    print("Anagrams")
else:
    print("No")

#or
from collections import Counter

if Counter(myStr1) == Counter(myStr2):
    print("Anagrams")
else:
    print("No")
```

Replace all occurrences of a substring in a string.

```python
fullText = "I am a good boy"
replacer = "bad"

fullText = fullText.replace("good", "bad")
print(fullText)

#specific occurences replacement
ot = "I am a very very bad boy"
replacero = ""

ot = ot.replace("very",replacero, 1) #Replaces single occurence
print(ot)
```

Count the frequency of each character in a string.

```python
from collections import Counter
text = "banana"
countedNum = Counter(text)
print(countedNum)
```

Given a string, print only the characters at even indices.

```python
myStr = "bangalore"
for i in range(1, len(myStr), 2):
    print(myStr[i])
```
