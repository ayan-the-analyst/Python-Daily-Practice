Reverse a three-digit number without using string conversion.

```python
def reverse_three_digit_number(num):
    if not (100 <= num <= 999):
        return "Input must be a three-digit number"

    hundreds = num // 100      
    tens = (num % 100) // 10   
    units = num % 10        

    reversed_num = (units * 100) + (tens * 10) + hundreds
    return reversed_num
num = int(input("Enter value: "))
print(f"The reverse of {num} is {reverse_three_digit_number(num)}")
```

Calculate the body mass index (BMI) given weight and height

```python
def bmiCal(h,w):
    return w/pow(h,2)

h = 1.2
w = 50
print(bmiCal(h,w))
```

Find the average of three numbers

```python
numList = []
for i in range(0,3):
    numIn = int(input(f"Enter value {i+1}: "))
    numList.append(numIn)
def getAvg(numList):
    return sum(numList)/3

print(f"The average value is : {getAvg(numList)}")
```
