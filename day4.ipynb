```python
a = 10
b = 80
c = 12
if a >=b:
    if a >= c:
        print(a)
    else:
        print(c)
else:
    if b >= c:
        print(b)
    else:
        print(c)
```

Print all numbers from 1 to N that are divisible by both 3 and 5.

```python
n = int(input("Enter value of N"))
for i in range (1,n+1):
    if i%3 == 0 and i%5 == 0:
        print(i,end=" ")
```

Check if a number is prime.

```python
num = 7
isPrime = True
for i in range (2,num):
    if num%i == 0:
        isPrime = False
        break

if isPrime:
    print("Prime")
else:
    print("Not prime")
```

Print the Fibonacci sequence up to N terms.

```python
n = 10
count = 0
n1, n2 = 0, 1
while count <= n:
    print(n1)
    nth = n1 + n2
    n1 = n2
    n2 = nth
    count += 1
```

Find the factorial of a number using a loop.

```python
n = 5
i = n
res = 1
while i > 0:
    res = res * i
    i -= 1
print(res)
```

Print a right-angled triangle pattern using asterisks.

```python
n = 10
for i in range(1, n + 1):
    for j in range(i):
        print("*", end="")
    print()
```

Find the sum of all digits of a number.

```python
n = 101
res =  0
while n != 0:
    rem = n % 10
    res = res + rem
    n = int(n/10)
print(res)
```

Find the GCD (Greatest Common Divisor) of two numbers.

```python
import math

num1 = 60
num2 = 48

result = math.gcd(num1, num2)
print(f"The GCD is: {result}")
```

Convert a binary number (string) to decimal.

```python
n = "100101101"
binres = 0
power = 0
rn = n[::-1]
for i in rn:
    intrn = int(i)
    binmid = intrn * pow(2,power)
    binres = binres + binmid
print(binres)
```
