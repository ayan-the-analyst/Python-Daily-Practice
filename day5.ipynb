1. Print all Armstrong numbers between 1 and 1000.

```python
def is_armstrong(n):
    digits = str(n)
    power = len(digits)

    total_sum = sum(int(digit) ** power for digit in digits)

    return n == total_sum

for num in range(1, 1001):
    if is_armstrong(num):
        print(num)
```

2. Find the sum of the series: 1 + 1/2 + 1/3 + ... + 1/N.

```python
def sum_of_series(n):
    total_sum = 0.0
    
    for i in range(1, n + 1):
        total_sum += 1 / i
        
    return total_sum

n = 20
result = sum_of_series(n)
print(f"The sum of the series up to {n} terms is: {result}")
```

3. Print a diamond pattern of asterisks for a given odd N.

```python
def print_diamond(n):

    if n % 2 == 0:
        print("Please provide an odd number.")
        return


    half = n // 2 


    for i in range(half + 1):
        spaces = " " * (half - i)
        stars = "*" * (2 * i + 1)
        print(spaces + stars)


    for i in range(half - 1, -1, -1):
        spaces = " " * (half - i)
        stars = "*" * (2 * i + 1)
        print(spaces + stars)


print_diamond(9)
```

7. Print the first N rows of Pascal's triangle.

```python
def print_pascals_triangle(n):

    triangle = []
    
    for i in range(n):
       
        row = [1] * (i + 1)
        
    
        for j in range(1, i):
          
            row[j] = triangle[i-1][j-1] + triangle[i-1][j]
            
       
        triangle.append(row)
        
   
    bottom_row_str = "   ".join(map(str, triangle[-1]))
    max_width = len(bottom_row_str)
    
    for row in triangle:
        row_str = "   ".join(map(str, row))
        
        print(row_str.center(max_width))


print_pascals_triangle(6)
```

8. Simulate a simple ATM menu (deposit, withdraw, check balance).

```python
def run_atm():
    
    balance = 1000.00  
    
    while True:
        
        print("\n" + "="*25)
        print("   🏦 SIMPLE ATM MENU   ")
        print("="*25)
        print("1. Check Balance")
        print("2. Deposit Money")
        print("3. Withdraw Money")
        print("4. Exit")
        print("="*25)
        
       
        choice = input("Please select an option (1-4): ")
        
        
        if choice == '1':
            print(f"\n💵 Your current balance is: ${balance:.2f}")
            
        
        elif choice == '2':
            amount_str = input("\nEnter amount to deposit: $")
            # A simple check to ensure the user typed a number (mostly)
            if amount_str.replace('.', '', 1).isdigit(): 
                amount = float(amount_str)
                if amount > 0:
                    balance += amount
                    print(f"✅ Successfully deposited ${amount:.2f}.")
                else:
                    print("❌ Please enter a positive amount.")
            else:
                print("❌ Invalid input. Please enter numbers only.")
                
        
        elif choice == '3':
            amount_str = input("\nEnter amount to withdraw: $")
            if amount_str.replace('.', '', 1).isdigit():
                amount = float(amount_str)
                if amount > balance:
                    print("❌ Insufficient funds! Your balance is lower than the requested amount.")
                elif amount > 0:
                    balance -= amount
                    print(f"✅ Successfully withdrew ${amount:.2f}.")
                else:
                    print("❌ Please enter a positive amount.")
            else:
                print("❌ Invalid input. Please enter numbers only.")
                
      
        elif choice == '4':
            print("\n👋 Thank you for using the ATM. Have a great day!")
            break  
            
       
        else:
            print("\n❌ Invalid choice. Please select 1, 2, 3, or 4.")


run_atm()
```

9. Find the roots of a quadratic equation.

```python
import cmath

def find_roots(a, b, c):
    
    if a == 0:
        return "Invalid input: 'a' cannot be 0."

    
    discriminant = (b**2) - (4 * a * c)

   
    root1 = (-b - cmath.sqrt(discriminant)) / (2 * a)
    root2 = (-b + cmath.sqrt(discriminant)) / (2 * a)

    return root1, root2


a = 1
b = -5
c = 6

r1, r2 = find_roots(a, b, c)


print(f"The roots of {a}x^2 + {b}x + {c} = 0 are:")
print(f"Root 1: {r1.real:.2f} + {r1.imag:.2f}j")
print(f"Root 2: {r2.real:.2f} + {r2.imag:.2f}j")
```

10. Implement a number guessing game (computer picks, user guesses).

```python
import random

def play_guessing_game():
  
    secret_number = random.randint(1, 100)
    attempts = 0
    
    print("\n" + "="*35)
    print(" 🎮 THE NUMBER GUESSING GAME 🎮 ")
    print("="*35)
    print("I am thinking of a number between 1 and 100.")
    
  
    while True:
        user_input = input("\nEnter your guess: ")
        
      
        if not user_input.isdigit():
            print("❌ Invalid input! Please enter a whole number.")
            continue 
            
       
        guess = int(user_input)
        attempts += 1
        
      
        if guess < secret_number:
            print("📉 Too low! Try aiming a little higher.")
        elif guess > secret_number:
            print("📈 Too high! Bring it down a notch.")
        else:
           
            print(f"🎉 CONGRATULATIONS! You guessed it in {attempts} attempts!")
            break 


# play_guessing_game()
```
